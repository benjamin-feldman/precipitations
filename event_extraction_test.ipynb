```python
def plot_time_series_with_events(ts, events):
    plt.figure(figsize=(12, 6))
    plt.plot(ts, label='Time Series',)  # Plot the time series data
    
    # Highlight each event with a shaded area
    for start, end in events:
        plt.axvspan(start, end, color='yellow', alpha=0.3)
    
    # Adding labels and title for clarity
    plt.xlabel('Time (*5 minutes)')
    plt.ylabel('Value')
    plt.title('TS avec événement segmenté')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def read_data(year, month, day):
    path = f'data/raw_data/{year:04d}/'
    file_name = f'RR_IDF300x300_{year:04d}{month:02d}{day:02d}.npy'
    full_path = path + file_name
    try:
        raw_data = np.load(full_path) / 100.0
        raw_data[raw_data < 0] = np.nan
        #raw_data[raw_data < 0] = 0 # ça changerait quoi ?
    except FileNotFoundError:
        raise FileNotFoundError(f"The file {full_path} does not exist.")
    return raw_data

def read_data_raww(year, month, day):
    path = f'data/raw_data/{year:04d}/'
    file_name = f'RR_IDF300x300_{year:04d}{month:02d}{day:02d}.npy'
    full_path = path + file_name
    try:
        raw_data = np.load(full_path) / 100.0
        #raw_data[raw_data < 0] = np.nan
        #raw_data[raw_data < 0] = 0 # ça changerait quoi ?
    except FileNotFoundError:
        raise FileNotFoundError(f"The file {full_path} does not exist.")
    return raw_data
```

```python
raw_data = read_data(2021, 1, 10)
ts = np.zeros((288))
for i in range(300):
    for j in range(300):
        ts += raw_data[:, i, j]

# Use the previously defined function to segment events
event_tuples = segment_events(ts)

# Plot the time series with highlighted events
plot_time_series_with_events(ts, event_tuples)
```

```python
def event_stats(raw_data, i, j, start_time, duration):
    events = raw_data[start_time:start_time+duration, i, j]
    events = events[~np.isnan(events)]
    return duration, np.max(events), np.mean(events), np.var(events), 1 - np.count_nonzero(events) / duration
```
